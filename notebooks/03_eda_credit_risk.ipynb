# 03 Exploratory Data Analysis

EDA on the real Zindi training table enriched with demographics and
historical-loan features. Focus on target balance, segment risk rates,
and data quality checks.

```python
%pip install -q pandas numpy
import sys, pathlib
sys.path.insert(0, str(pathlib.Path('..').resolve()))

import pandas as pd
from src.data import load_enriched_raw_competition_tables

ZIP_PATH = str(pathlib.Path('..') / 'data-science-nigeria-challenge-1-loan-default-prediction20250307-26022-im3qg9.zip')
train, test = load_enriched_raw_competition_tables(ZIP_PATH)
print(f"Train : {train.shape}  |  Test : {test.shape}")
print(f"Colonnes : {list(train.columns)}")
```

```python
print("=== TARGET ===")
vc = train['target_risk_flag'].value_counts()
print(f"Bad  (1) : {vc.get(1, 0):5d}  ({vc.get(1, 0)/len(train):.1%})")
print(f"Good (0) : {vc.get(0, 0):5d}  ({vc.get(0, 0)/len(train):.1%})")

print("\n=== DEMOGRAPHICS (bank_account_type) ===")
if 'bank_account_type' in train.columns:
    print(train.groupby('bank_account_type')['target_risk_flag'].agg(['count', 'mean']).rename(
        columns={'count': 'n', 'mean': 'taux_defaut'}
    ).sort_values('taux_defaut', ascending=False))

print("\n=== DISTRIBUTION SEXE ===")
if 'gender' in train.columns:
    print(train.groupby('gender')['target_risk_flag'].agg(['count', 'mean']).rename(
        columns={'count': 'n', 'mean': 'taux_defaut'}
    ))
```
