```python
%pip install -q pandas numpy
import sys, pathlib
sys.path.insert(0, str(pathlib.Path('..').resolve()))

from src.data.raw_competition import load_raw_competition_zip, combine_demographics

ZIP_PATH = str(pathlib.Path('..') / 'data-science-nigeria-challenge-1-loan-default-prediction20250307-26022-im3qg9.zip')
bundle = load_raw_competition_zip(ZIP_PATH)
print("Bundle chargé avec succès.")
```

```python
import pandas as pd

tables = {
    "train_demographics": bundle.train_demographics,
    "train_perf":         bundle.train_perf,
    "test_demographics":  bundle.test_demographics,
    "test_perf":          bundle.test_perf,
    "train_prevloans":    bundle.train_prevloans,
    "test_prevloans":     bundle.test_prevloans,
}
for name, df in tables.items():
    print(f"\n{'='*50}")
    print(f"  {name} — {df.shape[0]:5d} lignes × {df.shape[1]:2d} colonnes")
    print(f"{'='*50}")
    for col in df.columns:
        dtype = str(df[col].dtype)
        n_null = int(df[col].isnull().sum())
        null_str = f"  [{n_null} NaN]" if n_null else ""
        print(f"    {col:<40s}  {dtype:<10}{null_str}")
```
